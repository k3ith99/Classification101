```python
#pull data from csv and clean it
crypto_dict = {}
my_cryptos = ["BTC","ETH","LTC","BCH","XRP"]
def clean_data():
    for crypto in my_cryptos:
        csv = pd.read_csv(f"{crypto}.csv", header= 1)
        df2 = pd.DataFrame(csv)
        df2.index = pd.to_datetime(df2['date'])
        df3 = pd.DataFrame(df2['close'])
        df3['volume_usd'] = df2['Volume USD'].astype(int)
        daily_pct_change = df3['close'].pct_change()
        df3['daily_pct_change'] = daily_pct_change
        weekly_pct_change = df3['close'].pct_change(periods = 5)
        df3["weekly_pct_change"] = weekly_pct_change
        crypto_dict[f"{crypto}"] = df3
    
```

```python
clean_data()
```

```python
def closing_price():
    fig, ax = plt.subplots(figsize = (20,10))
    for crypto in my_cryptos:
        sns.lineplot(x='date', y='close', data=crypto_dict[crypto].loc['2018':], ax=ax,label = crypto)
    ax.set_ylabel("close")
    ax.set_yscale('log')
    ax.set_title('Closing Prices for Cryptocurrencies since 2018')
    sns.set_style("whitegrid")
    sns.set_context("paper")
    plt.tight_layout()
    ax.legend()
    plt.show()
```

```python
#simple moving average for 20 days
def moving_average():
    fig, ax = plt.subplots(figsize = (20,10))
    for crypto in my_cryptos:
        sns.lineplot(x='date', y='close', data=crypto_dict[crypto].loc['2018':].rolling(20).mean(), ax=ax,label = crypto)
    ax.set_ylabel("close")
    ax.set_yscale('log')
    ax.set_title('20 days rolling average for Cryptocurrencies since 2018')
    sns.set_style("whitegrid")
    sns.set_context("paper")
    plt.tight_layout()
    ax.legend()
    plt.show()
```

```python
def expo_moving_average():
    fig, ax = plt.subplots(figsize = (20,10))
    for crypto in my_cryptos:
        sns.lineplot(x='date', y='close', data=crypto_dict[crypto].loc['2018':].ewm(span = 20).mean(), ax=ax,label = crypto)
    ax.set_ylabel("close")
    ax.set_yscale('log')
    ax.set_title('20 days rolling exponentially weighed average for Cryptocurrencies since 2018')
    sns.set_style("whitegrid")
    sns.set_context("paper")
    plt.tight_layout()
    ax.legend()
    plt.show()
```

```python
def rolling_std():
    fig, ax = plt.subplots(figsize = (20,10))
    for crypto in my_cryptos:
        sns.lineplot(x='date', y='close', data=crypto_dict[crypto].loc['2018':].rolling(20).std(), ax=ax,label = crypto)
    ax.set_ylabel("close")
    ax.set_yscale('log')
    ax.set_title('Rolling standard deviation of Cryptocurrencies since 2018')
    sns.set_style("whitegrid")
    sns.set_context("paper")
    plt.tight_layout()
    ax.legend()
    plt.show()
```

```python
#boxplot of daily percentage changes
def box_plot_daily():
    df_daily = pd.concat([i['daily_pct_change'] for i in crypto_dict.values()], axis=1, keys=crypto_dict.keys())
    fig, ax = plt.subplots(figsize = (20,10))
    sns.boxenplot(data=df_daily,showfliers=False)
    plt.xlabel('Cryptocurrency')
    plt.ylabel('Daily Percentage Change')
    plt.title('Boxen Plots of Daily Percentage Change for Cryptocurrencies')
    sns.set_style("whitegrid")
    sns.set_context("paper")
    plt.tight_layout()
    plt.show()
```

```python
def box_plot_weekly():
    df_weekly = pd.concat([i['weekly_pct_change'] for i in crypto_dict.values()], axis=1, keys=crypto_dict.keys())
    fig, ax = plt.subplots(figsize = (20,10))
    sns.boxenplot(data=df_weekly,showfliers=False)
    plt.xlabel('Cryptocurrency')
    plt.ylabel('Weekly Percentage Change')
    plt.title('Boxen Plots of weekly Percentage Change for Cryptocurrencies')
    sns.set_style("whitegrid")
    sns.set_context("paper")
    plt.tight_layout()
    plt.show()
```

```python
#correlation matrix heatmap between all cryptocurrencies
def corr_matrix():
    df_close= pd.concat([i['close'] for i in crypto_dict.values()], axis=1, keys=crypto_dict.keys())
    sns.heatmap(df_close.corr(),annot = True, vmin = -1, vmax = 1)
    plt.show()
```
